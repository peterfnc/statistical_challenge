```python
data_out = data.pivot_table(index = 'OutcomeType', aggfunc = 'count')

data_out['probability'] = data_out['Color']/len(data['Color'])

data_out
```

```python
data_cat = data[data['AnimalType'] == 'Cat']

data_catout = data_cat.pivot_table(index = 'OutcomeType', aggfunc = 'count')

data_catout['probability'] = data_catout['Color']/len(data_cat['Color'])

data_catout
```

```python
data_dog = data[data['AnimalType'] == 'Dog']

data_dogout = data_dog.pivot_table(index = 'OutcomeType', aggfunc = 'count')

data_dogout['probability'] = data_dogout['Color']/len(data_dog['Color'])

data_dogout
```

```python
#no ocurre el mismo con gatos y perros. 
#estos datos no se deben a los proporciones de gatos y perros porque miramos a probabilidad
```

```python
data_sub = data.pivot_table(index = 'OutcomeSubtype', aggfunc = 'count')

data_sub['probability'] = data_sub['Color']/len(data['Color'])

data_sub_dist = data_sub['probability']

data_sub_dist = data_sub_dist.reset_index()
```

```python
import seaborn as sn
```

```python
import seaborn as sns
sns.set(style="whitegrid")
ax = sns.barplot(x="OutcomeSubtype", y="probability", data=data_sub_dist)
```

```python
data_age = data.pivot_table(index = 'AgeuponOutcome', aggfunc = 'count')

data_age['probability'] = data_age['Color']/len(data['Color'])

data_age_dist = data_age['probability']

data_age_dist = data_age_dist.reset_index()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
ax = sns.barplot(x="AgeuponOutcome", y="probability", data=data_age_dist)
ax.set_xticklabels(rotation=90)
```
